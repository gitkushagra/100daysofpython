```python
## COFFEE MAKER -- MAIN PROJECT
MENU = {
    "espresso": {
        "ingredients": {
            "water": 50,
            "coffee": 18,
        },
        "cost": 1.5,
    },
    "latte": {
        "ingredients": {
            "water": 200,
            "milk": 150,
            "coffee": 24,
        },
        "cost": 2.5,
    },
    "cappuccino": {
        "ingredients": {
            "water": 250,
            "milk": 100,
            "coffee": 24,
        },
        "cost": 3.0,
    }
}


condition = True
current = {"water": 300, "milk": 200, "coffee": 100, "money": 0}
while condition:
    choice = input("What would you like? (espresso/latte/cappuccino):")
    if choice == "report":
        print("Water: {water}ml\nMilk : {milk}ml\nCoffee: {coffee}g\nMoney: ${money}".format(**current))
    elif choice == "espresso":
        print("Please enter coins.")
        if current["water"] < MENU["espresso"]["ingredients"]["water"]:
            print("Sorry! There is not enough water.")
            continue
        if current["coffee"] < MENU["espresso"]["ingredients"]["coffee"]:
            print("Sorry! There is not enough coffee.")
            continue
        quarters = float(input("How many quarters? "))
        dimes = float(input("How many dimes? "))
        nickles = float(input("How many nickles? "))
        pennies = float(input("How many pennies? "))
        money = quarters*0.25+dimes*0.1+nickles*0.05+pennies*0.01
        if money < MENU["espresso"]["cost"]:
            print("Sorry, that's not enough money. Money refunded.")
            continue
        elif money >= MENU["espresso"]["cost"]:
            if money > MENU["espresso"]["cost"]:
                diff = money - MENU["espresso"]["cost"]
                print(f"Here is $"+str(round(diff,2))+" dollars in change.")
                money -= diff
            current["money"] += money
            current["water"] -= MENU["espresso"]["ingredients"]["water"]
            current["coffee"] -= MENU["espresso"]["ingredients"]["coffee"]
            print("Here is your Espresso. Enjoy!")
    elif choice == "latte":
        print("Please enter coins.")
        if current["water"] < MENU["latte"]["ingredients"]["water"]:
            print("Sorry! There is not enough water.")
            continue
        if current["milk"] < MENU["latte"]["ingredients"]["milk"]:
            print("Sorry! There is not enough milk.")
            continue
        if current["coffee"] < MENU["latte"]["ingredients"]["coffee"]:
            print("Sorry! There is not enough coffee.")
            continue
        quarters = float(input("How many quarters? "))
        dimes = float(input("How many dimes? "))
        nickles = float(input("How many nickles? "))
        pennies = float(input("How many pennies? "))
        money = quarters*0.25+dimes*0.1+nickles*0.05+pennies*0.01
        if money < MENU["latte"]["cost"]:
            print("Sorry, that's not enough money. Money refunded.")
            continue
        elif money >= MENU["latte"]["cost"]:
            if money > MENU["latte"]["cost"]:
                diff = money - MENU["latte"]["cost"]
                print(f"Here is $"+str(round(diff,2))+" dollars in change.")
                money -= diff
            current["money"] += money
            current["water"] -= MENU["latte"]["ingredients"]["water"]
            current["coffee"] -= MENU["latte"]["ingredients"]["coffee"]
            current["milk"] -= MENU["latte"]["ingredients"]["milk"]
            print("Here is your latte. Enjoy!")
    elif choice == "cappuccino":
        print("Please enter coins.")
        if current["water"] < MENU["cappuccino"]["ingredients"]["water"]:
            print("Sorry! There is not enough water.")
            continue
        if current["milk"] < MENU["cappuccino"]["ingredients"]["milk"]:
            print("Sorry! There is not enough milk.")
            continue
        if current["coffee"] < MENU["cappuccino"]["ingredients"]["coffee"]:
            print("Sorry! There is not enough coffee.")
            continue
        quarters = float(input("How many quarters? "))
        dimes = float(input("How many dimes? "))
        nickles = float(input("How many nickles? "))
        pennies = float(input("How many pennies? "))
        money = quarters*0.25+dimes*0.1+nickles*0.05+pennies*0.01
        if money < MENU["cappuccino"]["cost"]:
            print("Sorry, that's not enough money. Money refunded.")
            continue
        elif money >= MENU["cappuccino"]["cost"]:
            if money > MENU["cappuccino"]["cost"]:
                diff = money - MENU["cappuccino"]["cost"]
                print(f"Here is $"+str(round(diff,2))+" dollars in change.")
                money -= diff
            current["money"] += money
            current["water"] -= MENU["cappuccino"]["ingredients"]["water"]
            current["coffee"] -= MENU["cappuccino"]["ingredients"]["coffee"]
            current["milk"] -= MENU["cappuccino"]["ingredients"]["milk"]
            print("Here is your latte. Enjoy!")
            
```
